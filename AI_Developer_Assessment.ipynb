# Q2: Toy Machine Learning Classification Demo

For this demo, I used a Decision Tree Classifier in order to demonstrate how a ML model can learn to classify data into two groups.

## Steps:
1. Create a fake dataset:  
   - I'll use `make_classification()`. This gives us points with two features and two possible labels.
2. Split the data:  
   - I'll split the data into training and test sets. This will help us analyze how well the model is able to perform using new data.
3. Train a Decision Tree:  
   - I'll train the decision tree using the training data.
4. Predictions & Results:  
   - I'll make predictions on the test set. To evaluate the results, I'll print a classification report.
5. Plotting Results:  
   - Lastly, I'll plot the results to visualize how well the model works.

## Tools to Implement
- scikit-learn
- matplotlib
- seaborn

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Generate fake classification data
X, y = make_classification(n_samples=300, n_features=2, n_redundant=0, n_clusters_per_class=1, random_state=42)

# 2. Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Train a Decision Tree model
model = DecisionTreeClassifier(max_depth=3)
model.fit(X_train, y_train)

# 4. Make predictions and evaluate the model
y_pred = model.predict(X_test)
print("Classification Report:\n")
print(classification_report(y_test, y_pred))

# 5. Visualize results
plt.figure(figsize=(8,6))
sns.scatterplot(x=X_test[:, 0], y=X_test[:, 1], hue=y_pred, palette='coolwarm', s=60)
plt.title("Decision Tree Classifier Results")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```
